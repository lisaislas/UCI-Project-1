```python
# Dependencies and Setup
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
```

```python
# Pathing for clean data
clean_data = "clean_data/clean_food_data.csv"
# Read the cleaned data
cleaned_data_df = pd.read_csv(clean_data)
cleaned_data_df
```

```python
food_data_df = cleaned_data_df[["Gender","on_off_campus", "sports", "exercise", "cook"]]
food_data_df
```

```python
food_data_cooking_df = cleaned_data_df[["father_education", "mother_education", "parents_cook", "eating_out", "cook", "weight"]]
food_data_cooking_df
```

```python
male_df = food_data_df[food_data_df["Gender"] == 1]["exercise"].value_counts()
male_group = male_df.tolist()
female_df = food_data_df[food_data_df["Gender"] == 2]["exercise"].value_counts()
female_group = female_df.tolist()
```

```python
# Males 123 [34, 33, 8] no 45
# Females 123 [29, 16, 4] no 45
```

```python
X = np.arange(3)
plt.figure(figsize=(12,8))
plt.bar(X - .2, male_group, color = 'lightblue', edgecolor='black', width =.4)
plt.bar(X + .2, female_group, color = 'pink',edgecolor='black', width =.4)
plt.locator_params(axis='x', nbins=4)
plt.xlim(-.5,2.5)
plt.xticks(range(0,len(male_df.index)),["Everyday","Twice or Three Times per Week","Once a Week"])
plt.title("Gender Breakdown of Exercise Responses")
plt.xlabel("Response Type")
plt.ylabel("Number of Participants")
plt.legend(["Males", "Females"], loc="best", prop={'size': 15})
plt.savefig('Gender_Breakdown_of_Exercise_Responses.png')
plt.show()
```

```python
response_one = food_data_df[food_data_df["exercise"] == 1]
response_two = food_data_df[food_data_df["exercise"] == 2]
response_three = food_data_df[food_data_df["exercise"] == 3]
```

```python
pies = ["Yes","No"]
colors = ["Lightblue", "pink"]
exercise_no_filter = food_data_df["sports"].value_counts()
response_1 = response_one["sports"].value_counts()
response_2 = response_two["sports"].value_counts()
response_3 = response_three["sports"].value_counts()
fig, axs = plt.subplots(2,3,figsize=(15,15))
axs[0,0].axis('off')
axs[0,2].axis('off')
axs[0,1].pie(exercise_no_filter, labels=pies, colors=colors,
    autopct="%1.1f%%", shadow=True, startangle=90, radius=2)
axs[0,1].set_title("Do you play sports?")
axs[1,0].pie(response_1, labels=pies, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
axs[1,0].set_title("Exercise Everday")
axs[1,1].pie(response_2, labels=pies, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
axs[1,1].set_title("Exercise Twice or Three Times per Week")
axs[1,2].pie(response_3, labels=pies, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
axs[1,2].set_title("Exercise Once a Week")
plt.axis("equal")
plt.savefig('Sports_And_Excercise.png')
plt.show()
```

```python
food_data_df['on_off_campus'] = food_data_df['on_off_campus'].map({1:1, 2:2, 3:2, 4:2,})
plt.scatter(food_data_df["on_off_campus"], food_data_df["cook"], marker = "o",
            facecolors="pink", edgecolors="black", alpha=.75)
(slope, integer, rvalue, pvalue, stderr) = st.linregress(food_data_df["on_off_campus"], food_data_df["cook"])
regress_values = food_data_df["on_off_campus"] * slope + intercept
plt.locator_params(axis='x', nbins=2)
plt.xlim(0.5,2.5)
plt.xticks(range(1,3),["On Campus","Off Campus"])
plt.yticks(range(1,6),["Everyday","Few Times a Week","Sometimes","Only During Holidays","Never"])
line_eq = f'y= {str(round(slope,2))}x + {str(round(intercept,2))}'
plt.plot(food_data_df["on_off_campus"], regress_values, "r-")
plt.title("Living On/Off Campus & Cooking")
plt.xlabel("Living Situation")
plt.ylabel("Cooking")
print(f"The r-squared is: {rvalue}")
print(f"The equation of the regression line is: {line_eq}")
plt.savefig('Living_Situation_Cooking.png')
plt.show()
```

```python
col = food_data_cooking_df.loc[: , "father_education":"mother_education"]
food_data_cooking_df['average education of parents'] = col.mean(axis=1)
food_data_cooking_df
```

```python
plt.scatter(food_data_cooking_df['average education of parents'], food_data_cooking_df["eating_out"], marker = "o", facecolors="pink", edgecolors="black", s=food_data_cooking_df["eating_out"], alpha=.75)

(slope, intercept, rvalue, pvalue, stderr) = linregress(food_data_cooking_df['average education of parents'], food_data_cooking_df["eating_out"])

regress_values = food_data_cooking_df['average education of parents'] * slope + intercept
plt.xticks(range(1,6),["< high school","high school", "some college", "< college", "graduate degree"])
plt.yticks(range(1,6),["Never","1-2 times","2-3 times","3-5 times","everyday"])

line_eq = f'y= {str(round(slope,2))}x + {str(round(intercept,2))}'

plt.plot(food_data_cooking_df['average education of parents'], regress_values, "r-")
plt.title("Impact of parents' education on people's eating habit")
plt.xlabel("Parents' education")
plt.ylabel("Eating Out Frequency")

print(f"The r-squared is: {rvalue}")
print(f"The equation of the regression line is: {line_eq}")

plt.show()

#There is minimum relevance between a person's eating out habit and parents' education level. However, we can see that the higher level the parents' education is, the more often will people eat out
```

```python
plt.scatter(food_data_cooking_df['average education of parents'], food_data_cooking_df["parents_cook"], marker = "o", facecolors="red", edgecolors="black", s=food_data_cooking_df["eating_out"], alpha=.75)

(slope, intercept, rvalue, pvalue, stderr) = linregress(food_data_cooking_df['average education of parents'], food_data_cooking_df["parents_cook"])

regress_values = food_data_cooking_df['average education of parents'] * slope + intercept

plt.xticks(range(1,6),["< high school","high school", "some college", "< college", "graduate degree"])
plt.yticks(range(1,6),["Everyday","2-3 times/week","1-2 times/week","Only during holidays","never"])


line_eq = f'y= {str(round(slope,2))}x + {str(round(intercept,2))}'

plt.plot(food_data_cooking_df['average education of parents'], regress_values, "r-")
plt.title("Impact of parents' education on people's eating habit")
plt.xlabel("Parents' education")
plt.ylabel("Frequency of parents cooking at home")

print(f"The r-squared is: {rvalue}")
print(f"The equation of the regression line is: {line_eq}")

plt.show()

#Most parents cook at home almost everyday or at least 2-3 times a week regardless of the education level
```

```python
# data to plot
avg_parents_cook = food_data_cooking_df["parents_cook"].mean()
avg_eating_out = food_data_cooking_df["eating_out"].mean()
avg_self_cook = food_data_cooking_df["cook"].mean()

# create plot
fig, ax = plt.subplots()
bar_width = 0.2
X = np.arange(1)
plt.xticks(range(1,4),["parents cook","eating out","self cook"])
p1 = plt.bar(X, avg_parents_cook, bar_width, color='lightblue', label='parents cook')
p2 = plt.bar(X + bar_width, avg_eating_out, bar_width, color='pink', label='eating out')
p3 = plt.bar(X + bar_width * 2, avg_eating_out, bar_width, color='wheat', label='self cook')

plt.yticks(range(1,4),["Everyday","couple times/week","Not often"])

plt.xlabel('Eating habit')
plt.ylabel('Frequency')
plt.title('Eating habit frequency')

plt.legend()


plt.show()
```
